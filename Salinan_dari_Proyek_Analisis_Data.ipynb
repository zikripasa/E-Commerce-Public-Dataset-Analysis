### Cleaning Data

**Menangani Missing Value 1**

```python
order_reviews_df.isnull().sum()
```

```python
#Mengisi kolom yang kosong dengan "No Review"
order_reviews_df['review_comment_message'].fillna('No Review', inplace=True)
order_reviews_df['review_comment_title'].fillna('No Review', inplace=True)
```


**Menangani Missing Value 2**

```python
#Menangani Missing Values 2
orders_df.isnull().sum()
```

```python
orders_df.head()
```

Dikarenakan kolom yang memiliki missing values (order_approved_at, order_delivered_carrier_date, order_delivered_customer_date) berbentuk tanggal (datetime) dalam format object, maka langkah yang perlu dilakukan:

```python
#Konversi ke Date Time
orders_df[['order_purchase_timestamp', 'order_approved_at',
           'order_delivered_carrier_date', 'order_delivered_customer_date',
           'order_estimated_delivery_date']] = orders_df[
           ['order_purchase_timestamp', 'order_approved_at',
            'order_delivered_carrier_date', 'order_delivered_customer_date',
            'order_estimated_delivery_date']].apply(pd.to_datetime)
```

Jika order_approved_at kosong, bisa diasumsikan order belum disetujui kita isi dengan order_purchase_timestamp


```python
#Mengisi kolom order_approved_at dengan order_purchase_timestamp
orders_df['order_approved_at'].fillna(orders_df['order_purchase_timestamp'], inplace=True)
```

Jika order_delivered_carrier_date kosong, bisa diasumsikan pesanan belum dikirim oleh seller → Isi dengan estimasi waktu rata-rata antara purchase dan shipping

```python
#Mengisi kolom order_delivered_carrier_date dengan waktu rata-rata antara purchase dan shipping
avg_shipping_time = (orders_df['order_delivered_carrier_date'] - orders_df['order_purchase_timestamp']).median()
orders_df['order_delivered_carrier_date'].fillna(orders_df['order_purchase_timestamp'] + avg_shipping_time, inplace=True)
```

Jika order_delivered_customer_date kosong, pesanan mungkin belum sampai → Isi dengan estimasi median waktu pengiriman

```python
#Mnegisi kolom order_delivered_customer_date dengan median waktu pengiriman
avg_delivery_time = (orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']).median()
orders_df['order_delivered_customer_date'].fillna(orders_df['order_delivered_carrier_date'] + avg_delivery_time, inplace=True)
```

```python
#Hasil Akhir
orders_df.isnull().sum()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana hubungan antara jumlah cicilan pembayaran (payment_installments) dengan nilai pembayaran (payment_value), dan apakah ada perbedaan signifikan dalam pola pembayaran berdasarkan jenis pembayaran (payment_type)?

```python
# Scatter plot untuk melihat hubungan antara payment_installments dan payment_value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='payment_installments', y='payment_value', hue='payment_type', data=order_payments_df, alpha=0.6)
plt.title('Hubungan antara Jumlah Cicilan dan Nilai Pembayaran')
plt.xlabel('Jumlah Cicilan')
plt.ylabel('Nilai Pembayaran')
plt.legend(title='Jenis Pembayaran')
plt.show()

# Box plot untuk membandingkan pola pembayaran berdasarkan payment_type
plt.figure(figsize=(10, 6))
sns.boxplot(x='payment_type', y='payment_value', data=order_payments_df)
plt.title('Perbandingan Nilai Pembayaran berdasarkan Jenis Pembayaran')
plt.xlabel('Jenis Pembayaran')
plt.ylabel('Nilai Pembayaran')
plt.show()
```

### Pertanyaan 2:
Apakah produk dengan berat lebih tinggi (product_weight_g) dan dimensi fisik lebih besar (product_length_cm, product_height_cm, product_width_cm) cenderung memiliki biaya pengiriman (freight_value) yang lebih mahal, dan bagaimana pola ini bervariasi berdasarkan wilayah geografis penjual (seller_state)?

```python
# Merge data
merged_data = pd.merge(order_items_df, products_df, on="product_id")
merged_data = pd.merge(merged_data, sellers_df, on="seller_id")

# Visualisasi 1: Hubungan berat produk vs biaya pengiriman
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x="product_weight_g",
    y="freight_value",
    hue="seller_state",
    data=merged_data,
    palette="viridis",
    alpha=0.6,
    s=50
)
plt.title("Hubungan Berat Produk vs Biaya Pengirimen (Berdasarkan Negara Bagian Penjual)")
plt.xlabel("Berat Produk (gram)")
plt.ylabel("Biaya Pengiriman (R$)")
plt.legend(title="Negara Bagian Penjual", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

# Visualisasi 2: Korelasi dimensi produk dengan biaya pengiriman
dimensi_columns = ["product_length_cm", "product_height_cm", "product_width_cm", "freight_value"]
corr_matrix = merged_data[dimensi_columns].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Korelasi Dimensi Produk dengan Biaya Pengiriman")
plt.show()
```
